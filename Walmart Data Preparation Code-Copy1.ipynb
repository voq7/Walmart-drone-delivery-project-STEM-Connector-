```python
#Imporing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading the main data fie from the foder
data=pd.read_excel('/Users/vladimirant/Desktop/Walmart Project/WalmartCensusData/data_main.xlsx')
```

```python
#Checking the data
data
```

```python
#reading Median Famiy Income State Data
data_MedianIncomeState=pd.read_excel('/Users/vladimirant/Desktop/Walmart Project/WalmartCensusData/Data_MedianIncomeState.xlsx')
```

```python
#checking the MedianIncomeState data
data_MedianIncomeState
```

```python
#creating dictionary to merge tables by Median Family income by state
keys=data_MedianIncomeState['State']
values=data_MedianIncomeState['Median Income']
dictionary_MedianIncome = dict(zip(keys, values))
dictionary_MedianIncome
```

```python
#deleting extra-spaces in main data file in geographic areas names field
data['State']=data['State'].str.lstrip()
data['State']=data['State'].str.rstrip()
data['County']=data['County'].str.lstrip()
data['County']=data['County'].str.rstrip()
data['Geographic Area Name']=data['Geographic Area Name'].str.lstrip()
data['Geographic Area Name']=data['Geographic Area Name'].str.rstrip()
```

```python
#Adding MedianStateIncome column (Median Family Income)
data['MedianStateIncome'] = data.State.map(dictionary_MedianIncome)
```

```python
data
```

```python
#seecting rows where total population is not 0
data=data.loc[data['Total Population']!=0]
```

```python
#checking the data
data
```

```python
#creating the first list of the most important features/columns that need to be tarnsformed to numeric

List=['Total Population',
       'Population over 65',
       'Estimate!!Median income (dollars)!!FAMILIES!!Families',
       'MedianStateIncome']
```

```python
# Using this code all the variables that are not numerical in the columns 

#'Population over 65','Estimate!!Median income (dollars)!!FAMILIES!!Families',
#'MedianStateIncome', #'Total Population' will be deleted

for i in List:
    data=data[pd.to_numeric(data[i], errors='coerce').notnull()]
    
```

```python
#checking the data
data
```

```python
#deleting all the rows where 'Population over 65' is 0. 
data=data.loc[data['Population over 65']!=0]
data
```

```python
# creating the list of variables that also need to be transformed to numeric

List2=['% of population over 65',
       'Estimate!!Median income (dollars)!!FAMILIES!!Families',
       'Civilian noninstitutionalized population',
       'Population over 65 with disability',
       '%, Population over 65 with disability',
       'Civilian noninstitutionalized population with a disability',
       '% of Civilian noninstitutionalized population with a disability',
       'Population 25 years and over',
       'Population 25 years and over. High school graduate or higher',
       "Population 25 years and over!!Bachelor's degree or higher",
       '%, Population 25 years and over!!High school graduate or higher',
       "%, Population 25 years and over!!Bachelor's degree or higher",
       '%, Labor Force Participation Rate!!Population 16 years and over',
       '%, Labor Force Participation Rate!!Population 20 to 64 years',
       '%, Estimate!!Unemployment rate!!Population 16 years and over',
       '%, Estimate!!Unemployment rate!!Population 20 to 64 years']
```

```python
#running the code to transform to numeric
for i in List2:
    data=data[pd.to_numeric(data[i], errors='coerce').notnull().fillna('mean')]
```

```python
#checking the data
data
```

```python
#describing the data
data.describe()
```

```python
#transforming all the digit variables to numeric
data=data.apply(pd.to_numeric,errors='ignore')
data
```

```python
#Converting percentage columns to shares

data['% of population over 65']=data['% of population over 65']/100
data['%, Population 25 years and over!!Bachelor\'s degree or higher']=data['%, Population 25 years and over!!Bachelor\'s degree or higher']/100
data['%, Population 25 years and over!!High school graduate or higher']=data['%, Population 25 years and over!!High school graduate or higher']/100
data['%, Labor Force Participation Rate!!Population 16 years and over']=data['%, Labor Force Participation Rate!!Population 16 years and over']/100
data['%, Labor Force Participation Rate!!Population 20 to 64 years']=data['%, Labor Force Participation Rate!!Population 20 to 64 years']/100
data['%, Estimate!!Unemployment rate!!Population 16 years and over']=data['%, Estimate!!Unemployment rate!!Population 16 years and over']/100
data['%, Estimate!!Unemployment rate!!Population 20 to 64 years']=data['%, Estimate!!Unemployment rate!!Population 20 to 64 years']/100




```

```python
#Checking the data
data
```

```python
data.describe()
```

```python
#saving data on PC in excel format
data.to_excel("data_main_formated.xlsx",sheet_name='main_data')
```

```python
#creaing the table of census tracts qualifying to be low-icome areas
data_low_income_areas=data.loc[data['Estimate!!Median income (dollars)!!FAMILIES!!Families']<=data['MedianStateIncome']*0.8]
```

```python
#creating table of no low income census tracts (high income areas)
LowIncomeTracts_list=list(data_low_income_areas['id'])
data_high_income_areas=data[~data['id'].isin(LowIncomeTracts_list)]
data_high_income_areas
```
