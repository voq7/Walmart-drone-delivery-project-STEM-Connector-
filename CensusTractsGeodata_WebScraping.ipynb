```python
url_base='https://tigerweb.geo.census.gov/tigerwebmain/Files/tab20/tigerweb_tab20_tract_2020_'
```

```python
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY", "PR"]
```

```python
html_string=".html"
```

```python
URL_list=[]
for i in range(0,len(states)):
    url=url_base+states[i]+html_string
    URL_list.append(url)
    
    
```

```python
URL_list
```

```python
import pandas as pd
```

```python
TABLES=[]
for i in range (0, len(states)):
    table=pd.read_html(URL_list[i])[0]
    TABLES.append(table)
```

```python
Census_tract_geo=pd.concat(TABLES)
```

```python
Census_tract_geo
```

```python
Census_tract_geo["STATE"]=pd.to_numeric(Census_tract_geo["STATE"])
```
